```python
import torch
```

```python
x = torch.arange(4)
```

```python
x
```

```python
x = x.float()
x.requires_grad=True
```

```python
x.grad
```

```python
y = 2 * torch.dot(x,x)#y = 2x^2,对x求导为4x，所以1变化1，y变化4，2变化1，y变化8
```

```python
y
```

```python
y.backward()
x.grad #
```

```python
x.grad.zero_()
y = x.sum()
y.backward()
x.grad

```

```python
x.grad.zero_()
y = x*x
u = y.detach()#u一个常数
z = u*x
z.sum().backward()#对 z 的和反向传播。由于 u 常数，∂z/∂x = u，所以梯度直接拷贝 u。
x.grad == u
```
