```python
#totenser的使用
from PIL import Image   
from torch.utils.tensorboard import SummaryWriter
from torchvision import transforms

writer = SummaryWriter('./logs')
img = Image.open(r'/home/user1/.cache/kagglehub/datasets/thedatasith/hymenoptera/versions/5/cover.jpg')
print(img) #PIL类型

#totensor
trans_totensor = transforms.ToTensor()
img_tensor = trans_totensor(img)
writer.add_image('Tensor_img', img_tensor)

#normalize
print(img_tensor[0][0][0]) #查看第一个像素点的值
trans_normalize = transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
img_norm = trans_normalize(img_tensor)
print(img_norm[0][0][0]) #查看第一个像素点的值
writer.add_image('Normalize_img', img_norm, global_step=0) #添加到logs中
writer.close()
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("thedatasith/hymenoptera")

print("Path to dataset files:", path)
```

```python
trans_resize = transforms.Resize((224, 224))    
img_resize = trans_resize(img)
img_resize = trans_totensor(img_resize)
print(img_resize) #tensor
writer.add_image('Resize_img', img_resize, global_step=2)

```

```python
trans_resize2 = transforms.Resize(512)#定义一个工具箱，用于调整图像大小
```

```python
trans_compose = transforms.Compose([trans_resize2, trans_totensor]) # 定义一个工具箱，先调整大小，再转换为张量
#在compose中 数据需要是transforms类型
img_resize_2 = trans_compose(img)
writer.add_image('Resize2_img', img_resize_2, global_step=3)
```

```python
trans_randomcrop = transforms.RandomCrop(512) #定义一个工具箱，用于随机裁剪图像
trans_compose2 = transforms.Compose([trans_randomcrop, trans_totensor]) # 定义一个工具箱，先随机裁剪，再转换为张量
```

```python
for i in range(10):
    img_crop = trans_compose2(img)
    writer.add_image('RandomCrop_img', img_crop, global_step=i)
```

```python
writer.close()
```
